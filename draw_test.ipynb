```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon
import numpy as np
```

```python
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
```

```python
fig, ax = plt.subplots()

t = [5, 5, 10, 10, 0]
ax.add_patch(Rectangle((t[0] - t[2] / 2, t[1] - t[3] / 2), t[2], t[3],
             angle=t[4],
             rotation_point='center',
             edgecolor='blue',
             facecolor='lightblue',
             fill=True,
             alpha=0.5))


t = [13, 13, 10, 10, 44]
ax.add_patch(Rectangle((t[0] - t[2] / 2, t[1] - t[3] / 2), t[2], t[3],
             angle=t[4],
             rotation_point='center',
             edgecolor='blue',
             facecolor='lightblue',
             fill=True,
             alpha=0.5))

v = np.array([[8.94625, 10], [10, 10], [10, 8.90881]])
ax.add_patch(Polygon(v, closed=True, edgecolor='red', lw=2))

#display plot
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.show()
```

```python
x = v[:,0]
y = v[:,1]
print(PolyArea(x,y))
```
