```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon
import numpy as np
```

```python
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
```

```python
fig, ax = plt.subplots()

t = [-6, -3, 7, 7, 0]
ax.add_patch(Rectangle((t[0] - t[2] / 2, t[1] - t[3] / 2), t[2], t[3],
             angle=t[4],
             rotation_point='center',
             edgecolor='blue',
             facecolor='lightblue',
             fill=True,
             alpha=0.5))


t = [-6, -3, 7, 7, 50]
ax.add_patch(Rectangle((t[0] - t[2] / 2, t[1] - t[3] / 2), t[2], t[3],
             angle=t[4],
             rotation_point='center',
             edgecolor='blue',
             facecolor='lightblue',
             fill=True,
             alpha=0.5))

v = np.array([[-4.7261, 0.5], [-2.5, -1.36792], [-2.5, -4.2739], [-4.36792, -6.5], [-7.2739, -6.5], [-9.5, -4.63208], [-9.5, -1.7261], [-7.63208, 0.5]])
ax.add_patch(Polygon(v, closed=True, edgecolor='red', lw=2))

#display plot
pad_x = (np.max(v[:,0]) - np.min(v[:,0])) * 0.1
pad_y = (np.max(v[:,1]) - np.min(v[:,1])) * 0.1
plt.xlim(np.min(v[:,0]) - pad_x, np.max(v[:,0]) + pad_x)
plt.ylim(np.min(v[:,1]) - pad_y, np.max(v[:,1]) + pad_y)
plt.show()
```

```python
x = v[:,0]
y = v[:,1]
print(f"Area: {PolyArea(x,y)}")
```
