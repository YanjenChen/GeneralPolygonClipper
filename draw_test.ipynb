```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon
import numpy as np
```

```python
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
```

```python
fig, ax = plt.subplots()

ax.add_patch(Rectangle((1, 2), 8, 6,
             angle=0.0,
             rotation_point='center',
             edgecolor='blue',
             facecolor='lightblue',
             fill=True,
             alpha=0.5))

ax.add_patch(Rectangle((2, 2), 10, 10,
             angle=44.0,
             rotation_point='center',
             edgecolor='blue',
             facecolor='lightblue',
             fill=True,
             alpha=0.5))

v = np.array([[9, 2], [4.87763, 2], [1, 6.0154], [1, 8], [9, 8]])
ax.add_patch(Polygon(v, closed=True, edgecolor='red', lw=2))

#display plot
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.show()
```

```python
x = v[:,0]
y = v[:,1]
print(PolyArea(x,y))
```
